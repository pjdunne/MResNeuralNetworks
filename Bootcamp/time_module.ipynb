# Using the time module

`time` is a useful module in python for measuring runtime of code, along with many other use cases. Here we will make use of `time` to show the benefit of using `numpy.array` over a regular python `list` for large computations.

```python
# being by importing the relevant modules

%matplotlib widget
import numpy as np
import time
```

```python
# define our list and our numpy array

arr = np.linspace(1,1e8, int(1e8))
l = list(arr)
```

### +1 to every element: `list` vs `numpy.array`

In the cells below we will add 1 to every number in the array, measuring the time taken.

```python
# take the current time before and after finishing each task

print('Running using numpy array...')
t1_arr = time.time()
arr+=1
t2_arr = time.time()

print(f'Time taken using numpy array = {t2_arr - t1_arr}s\n')
print('Running using list...')
t1_list = time.time()
l = [i+1 for i in l]
t2_list = time.time()

print(f'Time taken using list        = {t2_list - t1_list}s')
```

The output from the above cell shows a clear disparity in computation time.

Below, we will plot runtime as a function of list/array length, to emphasise this more clearly.

```python
import matplotlib.pyplot as plt
```

```python
# this cell should take around 30s to run

lengths = []
arr_times = []
list_times = []

for i in range(7):

    length = 10**i
    lengths.append(length)
    arr = np.linspace(1,length, int(length))
    l = list(arr)
    
    t1_arr = time.time()
    arr+=1
    t2_arr = time.time()

    t1_list = time.time()
    l = [i+1 for i in l]
    t2_list = time.time()  

    arr_times.append(t2_arr - t1_arr)
    list_times.append(t2_list - t1_list)

plt.figure(figsize=(10,6))
plt.semilogx(lengths[:5], arr_times[:5], label='numpy array', lw=2)
plt.semilogx(lengths[:5], list_times[:5], label ='list', lw=2)
plt.xlabel('Length of Array')
plt.ylabel('Computation Time (s)')
plt.grid(which='both')
plt.legend()
plt.title('Up to list length of $10^4$')
plt.show()

plt.figure(figsize=(10,6))
plt.semilogx(lengths, arr_times, label='numpy array', lw=2)
plt.semilogx(lengths, list_times, label ='list', lw=2)
plt.xlabel('Length of Array')
plt.ylabel('Computation Time (s)')
plt.grid(which='both')
plt.legend()
plt.title('Up to list length of $10^8$')
plt.show()

```

These interactive plots compare the performance of the 2 data structures on increasingly large computations, and clearly show the superiority of numpy arrays.
