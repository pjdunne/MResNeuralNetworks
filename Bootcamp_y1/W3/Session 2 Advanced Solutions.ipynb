```python
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('Data/Fisher_iris.csv')
df.info()
df.head()
df.species.unique()
```

<div style="background-color: #00FF00">

** Exercise - Plot a histogram of the sepal_width for all 150 iris measurements. Now extract each species into a separate DataFrame and plot the sepal widths of each species on the same histogram to look at how this characteristic varies.**

```python
df['sepal_width'].hist()
```

```python
df_seto=df[df['species'].str.contains('setosa')]
df_vers=df[df['species'].str.contains('versicolor')]
df_virg=df[df['species'].str.contains('virginica')]

ax=df_seto['sepal_width'].hist(color='red',alpha=.3,label='setosa')
df_vers['sepal_width'].hist(color='blue',alpha=.3,ax=ax,label='versicolor')
df_virg['sepal_width'].hist(color='green',alpha=.3,ax=ax,label='virginica')
plt.legend()
```

# Seaborn Statistical Plotting

```python
import seaborn as sns
sns.set()
```

```python
import pandas as pd
df=pd.read_csv('Data/Fisher_iris.csv')
sns.distplot(df['sepal_width'],kde=False)
plt.show()
```

```python
help(sns.distplot)
```

<div style="background-color: #00FF00">

** Exercise - Using seaborn, plot each sepal_length of each species of iris as a histogram on the same axis. **

```python
sns.distplot(df_seto['sepal_length'],kde=False,label='Setosa')
sns.distplot(df_vers['sepal_length'],kde=False,label='Versicolor')
sns.distplot(df_virg['sepal_length'],kde=False,label='Virginica')
plt.legend()
```

## Linear Regression Plots

```python
sns.lmplot(x='sepal_length',y='sepal_width',data=df,fit_reg=False)
plt.show()
```
